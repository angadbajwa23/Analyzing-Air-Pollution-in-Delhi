```python
df.date=pd.to_datetime(df.date)
df=df.set_index('date')
df
```

```python
df=df.fillna({
    'stn_code':'0',
    'location_monitoring_station':'No station',
    'State':'Other',
    'location':'Other',
    'agency':'Unknown',
    'type':'Undetermined'
    
    })
df=df.interpolate(method='time')#very good for time index series
df.loc[:,'rspm']=df.loc[:,'rspm'].bfill()
df.loc[:,'pm2_5']=df.loc[:,'pm2_5'].bfill()# I had to use bfill as for thefirst few years there is no data for rspm and pm2_5(maybe they couldn't measure them at that point of time)

df
```

# Data_Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
pd.plotting.register_matplotlib_converters()
%matplotlib inline
g=subset_df.groupby('type')
for i,i_df in g:
    sns.scatterplot(x=i_df.index,y=i_df.so2,label=i)
    plt.title('SO2 variation across over 25 years')
    
    
```

```python
for i,i_df in g:
    print(i)
    sns.distplot(a=i_df.so2,label=i,kde=False)
    plt.title('SO2 variation across over 25 years')
```

```python
for i,i_df in g:
    print(i)
    sns.kdeplot(data=i_df.so2,label=i,shade=True)
    plt.title('SO2 variation across over 25 years')
```

```python
for i,i_df in g:
    print(i)
    sns.jointplot(y=i_df.so2,x=i_df.no2,label=i,kind='kde')
    plt.title('SO2 variation across over 25 years')
##SO2 AND NO2 TOGETHER
```
